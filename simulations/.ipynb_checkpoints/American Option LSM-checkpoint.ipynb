```python
%pylab inline --no-import-all
import numpy as np
import scipy as sp
import pylab as pl
import warnings 
import matplotlib.pyplot as plt
import statistics as st
from scipy.optimize import leastsq
import seaborn as sns
import pandas as pd
import math
```

Parameters: 

* $N $ : Number of paths we generate
* $m + 1$ : number of time steps
* $T$ Maturity 
* $\sigma$ volatility
* $r$ short return rate
* $q$ dividend 
* $S0$ Initial stock value
* $K$ Strike

Let's generate N stock paths following the SDE : 

$ \frac{dS_t}{S_t} = (r-q) dt + \sigma dW_t \quad $   (under risk-neutral probability)

```python
def generate_paths(S0, dt , sigma, r, N, m,q):
    ''' Function to generte stock paths.
    
    Parameters
    ==========
    S0    : float
            positive, initial Stock Value
    dt    : float
            time step discretization
    sigma : float 
            volatility 
    r     : float
            bond return 
    N     : int 
            Number of paths generated      
    m     : int 
            number of steps
    
    Returns
    =======
    S : matrix
        generated paths
    '''
    
    S = np.zeros((m + 1, N))
    S[0] = S0
    for t in range(1, m + 1):
        rand = np.random.standard_normal(N)
        S[t] = S[t - 1] * np.exp((r -q- sigma ** 2 / 2) * dt + sigma * dt ** 0.5 * rand)
    return S
```

We can plot a sample to see how it looks like 

```python
s =  generate_paths(100, 1./12 , 0.2, 0.03, 20, 11, 0.);
plt.plot(s[1:10]);
```

Now that we can generate paths, let's create the early exercise matrix. 

____

```python
def exercise_matrix(K,S): 
    
    ''' 
    Function to define the ealy exercise matrix.
    
    Parameters
    ==========
    S     : Matrix
            Stock Values 
    K     : float 
            Strike price
    
    Returns
    =======
    h : matrix
        early exercise for a call american option
    '''
    
    h=np.maximum(S-K,0)
    return h
```

_____


Let's compute the following dynamic pricing algorithm : 

$
\begin{equation} 
V_t = \max {(\mathbb{E}[e^{-r\Delta t} V_{t+\Delta t} | 
                                S_t], h_t)} \\
V_m = (S_m - K , 0)^+                              
\end{equation}
$

```python
def pricing_option(K,r,sigma,N,m,S_0,T,q,deg):
  
    "Time-step"
    dt=T/m
    "Discount factor"
    df=math.exp(-r*dt)
    "Creation of a stock path matrix"
    S=generate_paths(S_0, dt , sigma, r, N, m,q)
    "Creation of the exercise price matrix"
    h=exercise_matrix(K,S)
    "Initialization "
    V=h[-1]    

    "Iteration over the paths"
    for t in range (m-2,0,-1):
        reg = np.polyfit(S[t], V, deg)
        C = np.polyval(reg, S[t])
        V = np.where(h[t] > C, h[t], V * df)
    
    V_opt= df*np.mean(V)
    return (V_opt)
```

```python
import warnings
warnings.simplefilter('ignore', np.RankWarning) #The polynomial fit might induce some warnings 

T=1.
m=12
K=100.
S_0=100.
r=0.1
sigma=0.2
deg=5
N=100000
q=0.05

price=pricing_option(K,r,sigma,N,m,S_0,T,q,deg)
print ('price = ',price)
```
